```python
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python

url = 'https://tabelog.com/hyogo/rstLst/izakaya/'
restaurant_name =[]
kuchikomi_url = []
for j in range(50):
    taberogu = requests.get(url+str(j) + "/")
    print(url+str(j) + "/")
    soup = BeautifulSoup(taberogu.text, 'html.parser')
    for restauran in soup.find_all(class_ ='list-rst__rst-name-target cpy-rst-name'):
         print("店名",restauran.text)
         restaurant_name.append(restauran.text)  
         resturant_url = restauran.get("href")
         resturant_url = resturant_url + ("dtlrvwlst/")   
         print( resturant_url)
         kuchikomi_url .append( resturant_url)
    sleep(1)
```

```python
from janome.tokenizer import Tokenizer
from gensim.models.doc2vec import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
t = Tokenizer()
training_docs = []

kuchikomi_url 
review_texts ={}
for url ,name  in list(zip(kuchikomi_url,restaurant_name))[:3]:
    kuchikomis =requests.get(url)
    kuchikomis = BeautifulSoup(kuchikomis.text,'html.parser')
    kuchikomis = kuchikomis.find_all(class_='rvw-item__title-target')
    base_url = "https://tabelog.com"
    review_text   =""
    sleep(0.5)
    for kuchikomi in kuchikomis:
#       print(kuchikomi.get('href'))
        review = requests.get(base_url  + kuchikomi.get("href"))
        review = BeautifulSoup(review.text, 'html.parser')
        review_text += review.find(class_ ="rvw-item__rvw-comment").getText().replace('\n', "").replace('\u3000',"").replace('●',"")
        sleep(0.5)
    review_text =  t.tokenize(review_text,wakati=True)
    r = TaggedDocument(words =review_text,tags=[name])
    training_docs.append(r)
    print(name)
    
        
```

```python
print(training_docs)
```

```python
model = Doc2Vec(documents=training_docs, min_count=1, dm=0)
```

```python
print(model.docvecs['韓国料理 OMONI'])
print(model.docvecs.most_similar('韓国料理 OMONI'))
```
